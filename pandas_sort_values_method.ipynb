```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('matches - matches.csv')
```

```python
#yaha hua kya hai ki team1 and team2 column ko combine kiya hai 
#and unki values ko ascending order mey sort kar diya hai
(df['team1'].value_counts() + df['team2'].value_counts()).sort_values()
```

```python
#descending order mey print karane ke liye
(df['team1'].value_counts()+df['team2'].value_counts()).sort_values(ascending=False)
```

```python
#ascending order mey city data ko print karana ho to
df.sort_values('city')
```

```python
#descending order mey print karana ho to
df.sort_values('city',ascending=False)
```

```python
#ab maan lo jo aapne data mey  changes kiye unko permanent karna hai
#generally ye kaam nahi karna chahiye lekin agar karna ho to 
#inplace use karte hai

#by default inplace False hota hai magar
#hum usko True mey change kar sakte hai


#df.sort_values('city',ascending=False,inplace=true)
```

```python
#Ab tk humne kya kiya tha ki bhai ya to series data type ke upar sort function use kiya tha
#yaa dataframe mey kisi column ke upar sort function use kiya tha

#ab maan lo humey multiple columns ke upar sort function use karna hai
#In that case what we will do is sortvalues bracket ke andar list bana do 
#jo jo cloumn ko sort karna ho
df.sort_values(['city','date'])
```

```python
#Ab maan lo city ascending order mey chahiye aur date descending order mey chahiye
#Ek aur list bana di aur usme bata diya ki konse column ke liye kaise sorting chahiye
df.sort_values(['city','date'],ascending=[True,False])
```
